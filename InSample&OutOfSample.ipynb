##### 2). Train an ML model using SGDRegressor in the default setting.
Choose a random state number other than 30. Use this number for all random_state whenever random_state is a parameter.
The target column is 'W'. This column records the number of wins. The rest of the columns are characteristics of each player.  Exclude playerId, yearId, teamID, IgID from the features list.
Train the model with no normalization first. Copy the performance metrics to the hw6Template file for ease of grading. 
Then train the model again after normalization. Obtain both the out-of-sample and the in-sample performance.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn import metrics
```

```python
# Exclude 'playerID', 'yearID', 'teamID', 'lgID' from the features
X = df.drop(columns=['playerID', 'yearID', 'teamID', 'lgID', 'W'])  # Features
y = df['W']  # Target variable
```

```python
# Use a random_state other than 30, e.g., 41 as suggested
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=41)
```

```python
# Train the model without normalization
# Initialize untrain model with default parameter settings
model = SGDRegressor(random_state=41)
```

```python
#Train the model
model.fit(X_train, y_train)
```

```python
#Score the model
y_pred = model.predict (X_test)

#Evaluate the model

#r-squared indicates the proportion  of the variance in the dependant variable that is
# predictable from the independant variables in a regression model

mae = metrics. mean_absolute_error(y_test, y_pred)
mse = metrics. mean_squared_error (y_test, y_pred)
rmse = mse ** 0.5
r2 = metrics.r2_score(y_test, y_pred)

#Calculate Adjusted R-squared
n= len(y_test) # Number of observations
p = X_test.shape[1] # Number of predictors
adjusted_r2 = 1-(1-r2) * (n-1)/ (n - p - 1)

print(f"n is {n} and p is {p}.")

#print out the metrics
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squsred Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared : {r2}")
print(f"Adjusted R-squared: {adjusted_r2}")
```

```python
from sklearn.preprocessing import StandardScaler

# Normalize the data using StandardScaler
scaler = StandardScaler()
```

```python
# Fit the scaler on the training data and transform both X_train and X_test (without using pipeline)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 2. Training the model after normalization
model = SGDRegressor(random_state=41) 
model.fit(X_train_scaled, y_train)
```

```python
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)
```

```python
#Evaluate In-sample performance (on training data)
train_mae = metrics.mean_absolute_error(y_train, y_train_pred)
train_mse = metrics.mean_squared_error(y_train, y_train_pred)
train_rmse = train_mse ** 0.5
train_r2 = metrics.r2_score(y_train, y_train_pred)

#Calculate Adjusted R-squared
train_n= len(y_train) # Number of observations
train_p = X_train_scaled.shape[1] # Number of predictors
train_adjusted_r2 = 1-(1-train_r2) * (train_n-1)/ (train_n - train_p - 1)



# Evaluate Out-of-sample performance (on test data)
test_mae = metrics.mean_absolute_error(y_test, y_test_pred)
test_mse = metrics.mean_squared_error(y_test, y_test_pred)
test_rmse = test_mse ** 0.5
test_r2 = metrics.r2_score(y_test, y_test_pred)

#Calculate Adjusted R-squared
test_n= len(y_test) # Number of observations
test_p = X_test_scaled.shape[1] # Number of predictors
test_adjusted_r2 = 1-(1-test_r2) * (test_n-1)/ (test_n - test_p - 1)
```

```python
# Print the performance metrics for both training (in-sample) and test (out-of-sample) data
print(f"In-sample performance (on training data):")
print(f"Mean Absolute Error (MAE): {train_mae}")
print(f"Mean Squared Error (MSE): {train_mse}")
print(f"Root Mean Squared Error (RMSE): {train_rmse}")
print(f"R-squared (R²): {train_r2}")
print(f"Adjusted R-squared: {train_adjusted_r2}")

print(f"\nOut-of-sample performance (on test data):")
print(f"Mean Absolute Error (MAE): {test_mae}")
print(f"Mean Squared Error (MSE): {test_mse}")
print(f"Root Mean Squared Error (RMSE): {test_rmse}")
print(f"R-squared (R²): {test_r2}")
print(f"Adjusted R-squared: {test_adjusted_r2}")
```

##### 4) Train an ML model using RandomForestClassifier in the default setting.
Perform this transformation on 'W' and store the transformed value in 'W01' column: if W value is >10, assign 1 else assign 0. Use 'W01' as the label column and the same X without normalization as the feature columns. (Note, RandomForestClassifer is a tree-based algorithm and does not use regularization so there is no need to do normalization.) 
Choose a random state number other than 30. Use this number for all random_state whenever random_state is a parameter.
Train the model. Obtain both the out-of-sample and the in-sample performance.  Copy the performance metrics to the hw6Template file for ease of grading.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Create the W01 column based on the transformation
df['W01'] = df['W'].apply(lambda x: 1 if x > 10 else 0)
```

```python
X = df.drop(columns=['playerID', 'yearID', 'teamID', 'lgID', 'W', 'W01'])
y = df['W01']  # Label
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=41)
```

```python
# Initialize the RandomForestClassifier
rfc = RandomForestClassifier(random_state=41)
```

```python
# Train the model on the training data
rfc.fit(X_train, y_train)
```

```python
#score model with test set
y_pred = rfc.predict(X_test)

#Evaluate performance -- out of sample performance
print("OUT OF SAMPLE PERFORMANCE")
print("Model's Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Model's precision Score: ", metrics.precision_score(y_test, y_pred))
print("Model's recall: ", metrics.recall_score(y_test,y_pred))
print("Model's f1_Score: ", metrics. f1_score(y_test, y_pred))
```

```python
y_pred_insample = rfc.predict(X_train)

#Evaluate performance -- In Sample performance
print("IN-SAMPLE PERFORMANCE")
print("In-sample Accuracy: ", metrics.accuracy_score(y_train, y_pred_insample))
print("In-sample's precision Score: ", metrics.precision_score(y_train, y_pred_insample))
print("In-sample's recall: ", metrics.recall_score(y_train,y_pred_insample))
print("In-sample's f1_Score: ", metrics. f1_score(y_train, y_pred_insample))
```
