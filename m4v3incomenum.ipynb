```python
# read a csv file
df= pd.read_csv('incomenum.csv')
```

```python
#frequences of data label column
df['income_>50K'].value_counts()
```

```python
#put the feature column in dataframe x
#put the label column in y

x = df.drop(columns = ['income_>50K'], inplace = False) #features columns are all columns except the label column
y = df['income_>50K']
```

```python
#split data
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(x, y, test_size=0.3, random_state=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model= DecisionTreeClassifier (criterion="entropy", max_depth=4)
```

```python
model.fit(X_trainset,y_trainset)
```

```python
prediction=model.predict(X_testset)
```

```python
possibilities=model.predict_proba(X_testset)
```

```python
from sklearn import metrics
```

```python
print("Decision Tree's accuracy: ",metrics.accuracy_score(y_testset,prediction))
print("Decision Tree's precision score: ",metrics.precision_score(y_testset,prediction))
print("Decision Tree's recall: ",metrics.recall_score(y_testset,prediction))
print("Decision Tree's f1 score: ",metrics.f1_score(y_testset,prediction))
```
